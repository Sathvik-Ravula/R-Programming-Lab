## Generate Q-Q Plots For The DataSet

**The Q-Q Plot Is Generally Used To Detect Outliers In Data**

```r
data <- read.csv("9.csv")
```

```r
data
```

```r
names(data)
```

```r
data1 <- data[, -1]
```

```r
means <- colMeans(data1)
print(means)
```

```r
qqnorm(data1$EASTEU)
qqline(data1$EASTEU)
```

```r
qqnorm(data1$LATAM)
qqline(data1$LATAM)
```

```r
qqnorm(data1$INDIA)
qqline(data1$INDIA)
```

```r
qqnorm(data1$ENERGY)
qqline(data1$ENERGY)
```

```r
qqnorm(data1$MINING)
qqline(data1$MINING)
```

```r
qqnorm(data1$GOLD)
qqline(data1$GOLD)
```

```r
qqnorm(data1$WATER)
qqline(data1$WATER)
```

#### INTERPRETATION : 

From The Above Q-Q Plots, We Can See That the data is evenly distributed with very less number of outliers, especially in the columns LATAM, EASTEU (ONLY ONE OR TWO AWAY). 

For GOLD, ENERGY, INDIA, MINING we can see outliers along the tails.

##### WE CAN SEE THAT MOST OF THE COLUMNS IN DATA IS FOLLOWING THE MULTIVARIATE NORMAL DISTRIBUTION.

This result can be used to evaluate whether a data point may be an outlier and whether observed data may have a multivariate normal distribution. 

A Q-Q plot can be used to picture the *Mahalanobis* distances for the sample.
